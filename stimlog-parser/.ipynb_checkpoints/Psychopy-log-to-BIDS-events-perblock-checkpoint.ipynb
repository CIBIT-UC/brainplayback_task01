```python
# Create pandas dataframe to read data into data structure.
df_log=pd.read_csv(log_fn)

# Select subset of columns
df_log=df_log.loc[:,["MusicsSelect", "MusicExcerpt.started","baseline.started", "FixationCross.started"]]

# Remove rows with NaN - nor necessary for this event files.
df_log=df_log.dropna(how="any")
```

```python
df_log
```

```python

# Create pandas dataframe with columns required by the BIDS standard.
bids={'onset':[],'duration':[],'trial_type':[]}
df_bids=pd.DataFrame(bids)
```

```python
#number of decimals when rounding numbers
decimals=2

# offset of the experiment in seconds (seconds until first baseline.)
exp_offset=df_log.iloc[0]["baseline.started"]

# considering the log file, add rows for each event.

# For each row:
for i in range(len(df_log)-1):
    # Add Baseline trial.
    df_bids.loc[len(df_bids.index)] = [np.round(df_log.iloc[i]["baseline.started"]-exp_offset, decimals),
                                       np.round(df_log.iloc[i]["MusicExcerpt.started"]-df_log.iloc[i]["baseline.started"], decimals),
                                      'Baseline'] 
    
    df_bids.loc[len(df_bids.index)] = [np.round(df_log.iloc[i]["MusicExcerpt.started"]-exp_offset, decimals),
                                       np.round(df_log.iloc[i+1]["baseline.started"]-df_log.iloc[i]["MusicExcerpt.started"], decimals),
                                       df_log.iloc[i]["MusicsSelect"][-6:-4]] 
    
    
```
