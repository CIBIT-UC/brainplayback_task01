```python
import pandas as pd
import numpy as np

import os
```

```python
# Log file from psychopy.

log_fn=os.path.join('..',
                    'data', 
                    'logfiles_task01',                  
                    'sub-19_BrainplaybackQuadrants_2023-09-20_16h20.24.700_run-4.csv')

print(log_fn)
```

```python
#create a new lists by parsing the string using "," delimiter

head_tail = os.path.split(log_fn)
print(head_tail[1])

part_id=head_tail[1].split("_")

print("Participant: ",part_id[0])

run_id=part_id[-1].split(".")
print("Run id: ",run_id[0])
```

```python
results_fn = f"{part_id[0]}_music-list_{run_id[0]}.csv"

print(results_fn)
```

```python
# Create pandas dataframe to read data into data structure.
df_log=pd.read_csv(log_fn)
```

```python
df_log.head(20)
```

```python

# Select subset of columns
df_log_musics=df_log.loc[:,["musics"]]

df_log_musics = df_log.loc[df_log['musics'].str.contains("music", case=False, na=False)]

```

```python
df_log_musics
```

```python
df_log_musics=df_log_musics['musics'].str.replace('music', '../fMRI_stim/music')

df_log_musics=df_log_musics.str.replace( "\\" , "/")
```
