```python
import pandas as pd
import os
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

```

```python
# Load poms_results.tsv from behavioral folder
df = pd.read_csv(os.path.join('..','data','behavioral','poms_results.tsv'), sep='\t')

# Load ACC.tsv and add to df
df_acc = pd.read_csv(os.path.join('..','data','behavioral','ACC.tsv'), sep='\t')
df['accordance'] = df_acc['accordance']
df
```

```python
# remove sub-20
df = df[df['participant'] != 'sub-20']
```

```python
# find linear regression between accordance and poms total scores
slope, intercept, r_value, p_value, std_err = stats.linregress(df['accordance'],df['total'])
print('r_value: ', r_value)
print('p_value: ', p_value)
print('std_err: ', std_err)

# plot scatterplot with regression line
ax = sns.regplot(x='accordance', y='total', data=df)
ax.set(xlabel='Accordance', ylabel='POMS Total Score')

# add R_value, p_value to the plot
ax.text(0.05, 0.9, 'r_value: '+str(round(r_value,3)), transform=ax.transAxes)
ax.text(0.05, 0.85, 'p_value: '+str(round(p_value,3)), transform=ax.transAxes)

plt.show()
```
