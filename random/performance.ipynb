```python
# import data file
import pandas as pd
df = pd.read_csv('performance_results_allQs_predef.csv',sep=';')
```

```python
# remove last row of df (mean values) and show
df = df[:-1]
df
```

```python
# replace the Overall Acc values for subject 18, 19 and 20 with the average of Acc_Q1, Acc_Q2, Acc_Q3, Acc_Q4

df.loc[15,'Overall Acc'] = df.loc[15,'Acc_Q1':'Acc_Q4'].mean()
df.loc[16,'Overall Acc'] = df.loc[16,'Acc_Q1':'Acc_Q4'].mean()
df.loc[17,'Overall Acc'] = df.loc[17,'Acc_Q1':'Acc_Q4'].mean()

df
```

```python
# plot Acc_Q1, Acc_Q2, Acc_Q3, Acc_Q4, Overrall Acc and F1_score as a single boxplot
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Create an array with the colors you want to use (https://www.instagram.com/p/Cwzj6J0Ao-f/?img_index=6)
colors = ["#775106", "#B67D0E", '#D6A84E', '#EFE5D5', '#E1773B']

# Set your custom color palette
sns.set_palette(sns.color_palette(colors))

# draw line at y=0.25
plt.axhline(y=0.25, color='#B75C1B', linestyle='--', linewidth=1)

# draw line at x = 3.5
plt.axvline(x=3.5, color='k', linestyle=':', linewidth=1)

# boxplot
sns.boxplot(data=df[['Acc_Q1','Acc_Q2','Acc_Q3','Acc_Q4','Overall Acc']], 
            linewidth=1.5, fliersize=3, width=0.4, 
            medianprops={'linewidth': 2, 'color': '#B75C1B'}, 
            boxprops={'linewidth': 1.5}, 
            whiskerprops={'linewidth': 1.5}, 
            capprops={'linewidth': 1.5})

# show points
sns.swarmplot(data=df[['Acc_Q1','Acc_Q2','Acc_Q3','Acc_Q4','Overall Acc']], color="k", size=3.5)

# change x labels
plt.xticks([0,1,2,3,4], ['Q1','Q2','Q3','Q4', 'Overall'])

plt.ylabel('Test accuracy')
plt.show()
```
